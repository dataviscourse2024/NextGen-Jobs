```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('ai_job_market_insights.csv')

# Mapping dictionaries
ai_adoption_mapping = {'Low': 1, 'Medium': 2, 'High': 3}
automation_risk_mapping = {'Low': 1, 'Medium': 2, 'High': 3}
remote_friendly_mapping = {'Yes': 1, 'No': 0}
job_growth_mapping = {'Growth': 1, 'Decline': -1, 'Stable': 0}
# Applying the mappings to the respective columns
data['AI_Adoption_Level'] = data['AI_Adoption_Level'].map(ai_adoption_mapping)
data['Automation_Risk'] = data['Automation_Risk'].map(automation_risk_mapping)
data['Remote_Friendly'] = data['Remote_Friendly'].map(remote_friendly_mapping)
data['Job_Growth_Projection'] = data['Job_Growth_Projection'].map(job_growth_mapping)

# Display the updated dataset to check if the changes were applied correctly
data.head()
```

```python
output_file_path = '/Users/zhanyijun/Desktop/24Fall/NextGen-Jobs/processed_ai_job_market_insights.csv'
data.to_csv(output_file_path, index=False)
```
